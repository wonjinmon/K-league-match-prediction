# Test Data 만들기

 - 2023년 K리그 1의 36Rnd, 37Rnd 데이터를 가져와 모델에 맞게 전처리

```python
import pandas as pd
test_df = pd.read_excel('3637R.xlsx', engine='openpyxl')
test_df.head(3)
# display(test_df)
print(test_df.shape)
```

```python
# K_Columns 에서 컬럼명 가져오기
names = pd.read_excel('k_datasets\K_columns.xlsx', engine="openpyxl")
print(names.shape)
names.head()
```

```python
# 필요없는 열 제거 (성공%만 가져옴)
temp = names.drop(columns=['LABEL-45','LABEL-46','LABEL-47','LABEL-48']) 

# 컬럼명 가져와서 리스트로 저장
rows = temp.iloc[0].to_list()
print(len(rows))
print(rows)
```

```python
# df의 컬럼 개수를 K_columns에 맞춰 개수 맞추기
drop_columns = ['14', '15','20','21','23','24','27', '28' ,'30' ,'31' ,'33' ,'34', '36', '37', '39', '40','42', '43' ,'45',
               '46' ,'48', '49' ,'51' ,'52' ,'54', '55', '58', '59', '61', '62', '64', '65']
for idx in (drop_columns):
    # print(idx)
    test_df = test_df.drop(columns=[f'Unnamed: {int(idx)-1}'])

# 확인
print(test_df.shape)
# print(df.columns)
test_df.head(24)
```

```python
test_df.columns = rows
# test_test_df
```

```python
test_df['기준'] = None

a = ['ulsan','ulsan','suwon','suwon', 'pohang','pohang', 'jeju', 'jeju', 'jeonbuk','jeonbuk',
     'seoul', 'seoul', 'daejeon','daejeon', 'daegu','daegu', 'incheon', 'incheon', 'gangwon', 'gangwon',
     'gwangju', 'gwangju', 'suwonFC', 'suwonFC']

test_df['기준'] = a
test_df['시즌'] = 2023
```

```python
# 대회별 득점 비교 후 경기결과 열에 저장
for i in test_df["대회"].value_counts().index.tolist():
    for j in range(1,39): # 총 35개 라운드 수집
        H_score = test_df[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "HOME")]["득점"]
        A_score = test_df[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "AWAY")]["득점"]
        if len(H_score) > 0:
            if H_score.iloc[0] > A_score.iloc[0]:
                test_df.loc[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "HOME"), "경기결과"] = 'W'
                test_df.loc[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "AWAY"), "경기결과"] = 'L'
            elif H_score.iloc[0] < A_score.iloc[0]:
                test_df.loc[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "HOME"), "경기결과"] = 'L'
                test_df.loc[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "AWAY"), "경기결과"] = 'W'
            else:
                test_df.loc[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "HOME"), "경기결과"] = 'D'
                test_df.loc[(test_df["대회"] == i) & (test_df["Rnd."] == j) & (test_df["H/A"] == "AWAY"), "경기결과"] = 'D'
```

```python
# 보기좋게 라운드 순으로 정렬
test_df = test_df.sort_values('Rnd.')

# 36 라운드 승패 확인
# test_df.head(12)
```

# 트랜스퍼마켓 데이터 전처리

```python
tm_data_2023 = pd.read_excel('k_datasets/K_trsfmk_2023.xlsx', engine='openpyxl').drop(columns=['LABEL-1','LABEL-2','LABEL-4'])

# 열 이름 변경경
tm_data_2023 = tm_data_2023.rename(columns={'LABEL-3' : '구단', 'LABEL-5' : '소속 선수 수',
                                'LABEL-6' : '평균 나이','LABEL-7' : '용병 수',
                                'LABEL-8' : '평균 가치', 'LABEL-9' : '총 가치'})

# 구단 이름 수정
tm_data_2023['구단'] = ['jeonbuk', 'ulsan', 'seoul', 'suwon', 'daejeon', 'gangwon',
                           'jeju', 'incheon', 'gwangju', 'pohang','daegu','suwonFC']

# 단위 제거
tm_data_2023['평균 가치'] = tm_data_2023['평균 가치'].str.replace('€', '').str.replace('k', '').astype(float)
tm_data_2023['총 가치'] = tm_data_2023['총 가치'].str.replace('€', '').str.replace('m', '').astype(float)
```

```python
# df에 tm_data_2023 합치기
tm_columns = tm_data_2023.columns[1:]

# df에 열 먼저 생성
for column in tm_columns:
    test_df[f'{column}'] = None
# test_df.head(1)


#2023
for column in tm_data_2023.columns[1:]:
    for i in test_df['기준']:
        for j in tm_data_2023['구단']:
            if i == j:
                test_df.loc[(test_df['기준'] == i) &(test_df['시즌'] == 2023), f'{column}'] = tm_data_2023.loc[tm_data_2023['구단'] == j , f'{column}'].values[0]
print('2023 end !')

test_df.head(1)
```

```python
import pandas as pd

# 'df'라는 이름으로 DataFrame이 있다고 가정합니다.
# '대회', '시즌', '라운드', 'H/A'를 기준으로 각 대회에 대한 두 행을 찾아 합칩니다.

new_rows = []

# '대회', '시즌', '라운드', 'H/A' 열을 기준으로 각 대회에 대한 행을 찾아 합칩니다.
unique_competitions = test_df['대회'].unique()

for competition in unique_competitions:
    competition_rows = test_df[test_df['대회'] == competition]
    
    # '시즌', '라운드', 'H/A' 열을 고려하여 대회에 대한 행을 찾아 합칩니다.
    unique_seasons = competition_rows['시즌'].unique()
    
    for season in unique_seasons:
        season_rows = competition_rows[competition_rows['시즌'] == season]
        
        unique_rounds = season_rows['Rnd.'].unique()
        
        for rnd in unique_rounds:
            round_rows = season_rows[season_rows['Rnd.'] == rnd]
            
            # 'H/A' 열을 기준으로 원정과 원정를 구분하여 대회, 시즌, 라운드에 대한 두 행을 합칩니다.
            home_row = round_rows[round_rows['H/A'] == 'HOME'].iloc[0]
            away_row = round_rows[round_rows['H/A'] == 'AWAY'].iloc[0]
            
            merged_row = pd.concat([home_row, away_row]).reset_index(drop=True)
            new_rows.append(merged_row)

# 새로운 DataFrame을 생성합니다.
new_test_df = pd.DataFrame(new_rows)

# 새로운 DataFrame을 출력합니다.
new_test_df
```
