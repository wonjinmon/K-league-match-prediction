```python
import pandas as pd
import numpy as np

# 데이터셋 불러오기
df = pd.read_excel("k_datasets/2023_daegu.xlsx", engine="openpyxl")
print(df.shape)
df.head()
```

컬럼명 가져오기

```python
names = pd.read_excel("k_datasets/K_columns.xlsx", engine="openpyxl")
print(names.shape)
names.head()
```

```python
# 크기 맞춰주기
temp = names.drop(columns=['LABEL-45','LABEL-46','LABEL-47','LABEL-48'])
temp
```

```python
# 컬럼명 가져와서 리스트로 저장
rows = temp.iloc[0].to_list()
print(len(rows))
print(rows)
```

컬럼명 붙여넣기

```python
# 컬럼 개수를 먼저 맞춰준다
drop_columns = ['14', '15','20','21','23','24','27', '28' ,'30' ,'31' ,'33' ,'34', '36', '37', '39', '40','42', '43' ,'45',
               '46' ,'48', '49' ,'51' ,'52' ,'54', '55', '58', '59', '61', '62', '64', '65']
for idx in (drop_columns):
    df = df.drop(columns=[f'LABEL-{idx}'])
```

```python
# 확인
print(df.shape)
print(df.columns)
df.head()
```

```python
# 컬럼명 붙여넣기
df.columns = rows
df.tail(3)
```
