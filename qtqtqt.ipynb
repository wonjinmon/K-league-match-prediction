```python
from glob import glob

import pandas as pd


datas = glob('k_datasets/*.xlsx')

teams = [xlsx.split('_')[-2] for xlsx in datas[:12]]

df = pd.read_excel(f'{datas[0]}', engine='openpyxl')

df['기준'] = teams[0]

for xlsx in datas[1:12]:
    team_name = xlsx.split('_')[-2]
    df_temp = pd.read_excel(f'{xlsx}', engine="openpyxl")
    df_temp["기준"] = f'{team_name}'
    df = pd.concat([df, df_temp])
print(df.shape)
df.tail()
```

```python
df["기준"].value_counts()
```

```python
names = pd.read_excel(f'{datas[-1]}', engine="openpyxl")
print(names.shape)
names.head()
```
