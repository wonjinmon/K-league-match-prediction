# 최적변수 관련 설명

변수선택법에 3가지 종류가 있음

- 전진 선택법 : 기존 모형에 가장 설명력이 좋은 변수를 하나씩 추가해가는 방법

- 후진 소거법 : 모든 변수가 포함된 모형에서 설명력이 안좋은 변수를 하나씩 제거해 나가는 방법

- 단계적 선택법 : 전진 선택법에서 후진 소거법을 추가한 방법

일반적으로 단계적 선택법이 좋음. 키 페이퍼도 단계적 선택법을 사용하였음.

단계 선택법이란 ? 최소 p 값과 유의 수준, 최종적으로 최대 p 값까지 고려하여 변수를 추가/제거 하는 방법

## 1.

처음 수집한 데이터를 df로 만들었을 때, (1332, 52) 의 쉐잎이었음.

한 라운드의, 1개의 매치(컬럼명 :대회)에 2개의 행을 가지고 있는 형식

ex) 1라운드/  포항vs제주 /  - 포항 데이터 -

ex) 1라운드/  제주vs포항 /  - 제주 데이터 - 

따라서 대회 변수, H/A 변수를 더미변수로 만들었음

이 과정에서 더미변수가 너무 많아져버림. 구단과 구단 사이에 수많은 경우의 수가 0 or 1로 구성되었다.

이 데이터프레임 형식일때 모델의 정확도는 전체적으로 60% 정도

- 이 df로 변수 선택법 하니 225개중 50여개정도. 성능이 좋아지지도 않았음

------------

더미 변수가 너무 많아 학습력이 떨어진다고 생각했음 ( 0이 너무 많았다고 생각, 열도 너무 늘어났음 )

더미 변수를 줄이기 위해, 수집한 데이터의 형식을 바꿔버림

기존 한개의 라운드, 1개의 매치에 2개의 행을 가지고 있는 형식 >>> 한개의 라운드, 1개의 매치에 1개의 행을 가지고 있는 형식

ex) 1라운드 / 포항 vs 제주 / - 포항(홈) 데이터 - / - 제주(원정) 데이터 - 

이에 따라 데이터 쉐잎도 바뀜 ( 662,100 )  

더미 변수 개수가 이전엔 200여개였는데 100여개로 줄었음

이 데이터 프레임 형식일때의 모델의 정확도는 전체적으로 80% 이상이다.

- 이 df로 변수 선택법 하니 4개. 성능 오름
